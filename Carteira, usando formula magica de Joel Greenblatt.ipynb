```python
#importando as bibliotecas 

import numpy as np
import pandas as pd
import string
import warnings
import yfinance as yf
warnings.filterwarnings('ignore')

import requests
```

```python
#site usado para extrair dados foi fundamentus
url = 'http://www.fundamentus.com.br/resultado.php'
```

```python


header = {
  "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"  
 }

r = requests.get(url, headers=header)
```

```python
# começamos tratar dados, site traz decimal ponto e python é virgula e milhar ponto.

df = pd.read_html(r.text,  decimal=',', thousands='.')[0]
df
```

```python
# colunas que tem porcentagem transformar em decimal
for coluna in ['Div.Yield', 'Mrg Ebit', 'Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
  df[coluna] = df[coluna].str.replace('.', '')
  df[coluna] = df[coluna].str.replace(',', '.')
  df[coluna] = df[coluna].str.rstrip('%').astype('float') / 100
```

```python
df
```

```python
# Aplicar formula magíca
# eliminar ativos baixa liquidez 

maiorLiquidez = df[df['Liq.2meses'] >=1000000] 
maiorLiquidez
```

```python
# não pode ter dividas maior que patrimonio da empresa

patrimonioLiquido = maiorLiquidez[maiorLiquidez['Patrim. Líq']>0]
patrimonioLiquido
```

```python
# agora precisamos somente das colunas papel, ROIC e EV/ebitda

dataNovo = patrimonioLiquido[['Papel','ROIC', 'EV/EBITDA']]

dataNovo
```

```python
# classificamos o EV/EBITDA do menor para maior

EVordenado = dataNovo.sort_values(by = 'EV/EBITDA' )
EVordenado = EVordenado.reset_index(drop=True)
EVordenado['EVordenado']= EVordenado.index
EVordenado
```

```python
# classificamos o ROIC do maior para menor 

ROICordenado = EVordenado.sort_values(by = 'ROIC', ascending= False)
ROICordenado = ROICordenado.reset_index(drop=True)
ROICordenado['ROICordenado'] = ROICordenado.index
dados = ROICordenado
dados
```

```python
# soma-se as classificação, a soma final é organizada da menor para maior 

dados['score'] = dados['EVordenado']+ dados['ROICordenado']
dados.sort_values(by = 'score')
```

```python
# obtendo o setor das empresas selecionadas 

def GetSector(nomeDoAtivo):
    setor = yf.Ticker(nomeDoAtivo)
    return setor.info['sector']
```

```python
symbols = dados['Papel']+'.SA'

sectorDict = {}

for tick in symbols:
    try:
        print("coletando dados do ativo:", tick)
        sectorDict[tick] = GetSector(tick)
    except:
        print("Não deu certo para ativo:", tick)
        sectorDict[tick] = 'NA'
stockSectorDF = pd.DataFrame.from_dict(sectorDict, orient = 'index'). reset_index()
stockSectorDF.columns = ['Papel', 'sector']


stockSectorDF
```

```python
stockSectorDF
```

```python
# adicionamos .SA para juntar as tabelas 

dados['Papel'] = dados ['Papel']+ '.SA'
dados
```

```python
# juntamos as tabelas 

dadosNovos = pd.merge(dados, stockSectorDF, on='Papel', how = 'inner')
dadosNovos
```

```python
# trinta melhores empresas sem exclusão de setores 

dadosNovos.sort_values(by = 'score'). head(30)
```

```python
# quais setores temos 

dadosNovos['sector']. unique()
```

```python
# excluimos os setores finanças, energia, sem setor definido e prestadores serviço publico.

filtroSetores = dadosNovos[dadosNovos['sector'] != 'Financial Services' ]
filtroSetores = filtroSetores[filtroSetores['sector']!= 'Energy']
filtroSetores = filtroSetores[filtroSetores['sector']!= 'Utilities']
filtroSetores = filtroSetores[filtroSetores['sector']!= 'NA']
filtroSetores
```

```python
# aqui esta resultados das melhores açãoes segundo a formula. 

filtroSetores.sort_values(by = 'score'). head(30)
```
